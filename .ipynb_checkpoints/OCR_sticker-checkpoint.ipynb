```python
import cv2, os
import numpy as np
import matplotlib.pyplot as plt
import imutils
import ultralytics as ul
import easyocr, torch

from transformers import TrOCRProcessor, VisionEncoderDecoderModel
import requests
from PIL import Image

%matplotlib inline
```

```python
from ultralytics import YOLO

img_path = '/workspace/project_yolo_tama/runs/detect/predict5/crops/plate/IMG-20230816-WA0015.jpg'
img = cv2.imread(img_path)
#img = cv2.resize(img, (640,640))

plt.imshow(img)
```

```python
model = YOLO("/workspace/yolo_obj_detect_try/yolov8s_custom.pt")  # load a pretrained model
tmp = model.predict(img)
```

```python
class_ = 1  # -> Plate, 1 -> Sticker

try:
    class_indices = np.where(tmp[0].boxes.cls.cpu().numpy() == class_)[0]

    if 1 in class_indices:
        index_1_position = class_indices[0]
        tmp_arr = tmp[0].boxes.xyxy.cpu().numpy()[index_1_position]
        x_min = int(tmp_arr[0])
        y_min = int(tmp_arr[1])
        x_max = int(tmp_arr[2])
        y_max = int(tmp_arr[3])

        extracted_region = img[y_min:y_max, x_min:x_max]

        #plt.imshow(extracted_region)
        print("VIP")
    else:
        print("Umum")
        
except:
    print(f'Class : {class_} is not found')
```
